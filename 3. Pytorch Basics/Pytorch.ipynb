```python
# Difference between tensor, vectors, and scalar

# Scalar
# A scalar has magnitude (size) only and no direction.
# Examples:
# Temperature = 30°C
# Mass = 5 kg
# Time = 10 s
# Speed = 60 km/h
# If someone says "the temperature is 30°C", that is enough information. No direction is needed.

scalar = torch.tensor(5)
print(scalar)
print(scalar.shape)
```

```python
# 2. Vector
# A vector has magnitude and direction.
# Examples:
# Velocity
# Force
# Acceleration
# Displacement

vector = torch.tensor([1, 2, 3, 4])

print(vector)
print(vector.shape)
```

```python
# 3. Tensor - a tensor is a multi-dimensional array that generalizes scalars and vectors.
# A tensor can represent data in multiple dimensions at once.
# Scalar = single number
# Vector = list of numbers
# Tensor = table or higher-dimensional array of numbers
# Example: stress values in a solid object can be represented as a tensor.
# A tensor is represented as an n-dimensional matrix.

matrix = torch.tensor([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix.shape)
```

```python
#Torch creation tensor methods
random=torch.rand(3,4) #random tensor of size 3x4
print(f"rand: {random}")

zeros=torch.zeros(4,5)#tensor of specified size of all 0
print(f'zero: {zeros}')

ones=torch.ones(4,5)# tensor of specified size of all 1
print(f'ones: {ones}')

eye = torch.eye(3)# identity matrix of given size
print(f'eye: {eye}')

arange = torch.arange(start=0, end=10, step=2)
print(f'arange: {arange}')
```

```python
#Tensor Arithmetic

a = torch.tensor([1, 2, 3])
b = torch.tensor([3, 4, 5])

print(a+b)
print(a-b)
print(a*b)

# adding using add()
print(torch.add(a,b))

print(torch.add(a, b, alpha=2))  # steps (a + bx)

```

```python
# Element-wise multiplication (*)
# Element-wise multiplication multiplies values at the same index in each tensor.
# For example, position [0] in tensor A is multiplied with position [0] in tensor B.
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print(f"Tensor A: {a}")
print(f"Tensor B: {b}")
print(f"Element-wise multiplication (A * B): {a * b}")

# Matrix multiplication (@)
mat_a = torch.tensor([[1, 2],
                      [3, 4]])
mat_b = torch.tensor([[5, 6],
                      [7, 8]])

print(f"
Matrix A: {mat_a}")
print(f"Matrix B: {mat_b}")
print(f"Matrix multiplication (A @ B): {mat_a @ mat_b}")

print(f"Element-wise multiplication (A * B): {mat_a * mat_b}")
```

```python
# reshape vs view
# `reshape` returns a tensor with a new shape. It may allocate new memory if needed.
# `view` returns a tensor with the same data and requires the tensor to be contiguous in memory.
t = torch.arange(1, 10, 1)  # numbers from 1 to 9
print(t.reshape(3, 3))  # converting t to a 3x3 tensor
print(t.view(3, 3))  # view shares memory with the original tensor
print(t)
```
